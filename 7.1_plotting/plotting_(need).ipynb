# 1 A simple plot

## 1.1 Let’s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values', # only y_1 values were used for Y
         color='red', linestyle='dashed')
plt.xlabel('x-values') # horizontal axis label
plt.ylabel('y-values') # vertical axis label
plt.title('X vs Y') # name of graph
plt.grid(alpha=0.1) # opacity of grids in the background, values <= 1
plt.legend(loc='upper left') # loc = location
```

### Things to note

```python
# The first 5 lines of code above are essential for producing the graph,
# they specify the details of the graph and plots.
# However, the subsequent lines are more optional add-ons.

# Abbreviations:
# color = c, linestyle = ls, linewidth = lw

# add plt.show() at the very end of the code to produce your graph

# plt.plot(x, y_1, y_2) will produce 2 lines of (x, y_1) and (x, y_2)

# order of keywords in plt.plot() doesnt matter

# for plt.legend(); keywords are "upper" "lower" "right" "left"
# can name legend using title = 'name'
```

## 1.2 Adding another plot

```python
# original:
plt.plot(x, y_1, label= 'Y1 values', color='red', ls='dashed')

plt.plot(x, y_2,
         label='Y2 values', color='blue')

plt.legend(loc="lower right")
```

## 1.3 Yet another plot but with error bars

```python
y_3 = [0, 4, 8, 12, 16, 20, 24]

x_error, y_error = .1, 0.75 # defines CONSTANT error in x, y values 
# can provide a list of error points instead as well.

plt.errorbar(x, y_3, ## should include y_3 data set in this part of website..
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green")
```

# 2 Better with NumPy

```python
import numpy as np
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# converting PyLists to NpArrays
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

# usning np, make variables that convert x to x2 and sin(x):
x2 = x**2                
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )#?WHY is there a ", " at end?
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

- The plot does not show proper shape of sin(x) due to range of the y axis, where sin(x) lies between +-1 range while x2 has no such bounds.

## We need another axis!

### Things to note

```python
# Using twinx() to produce 2 y-axes sharing the same x-axis

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left') # For graph on y-axis 1, x2 graph

plt.twinx() # This creates a new y-axis for the plots that comes after

plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right') # For graph on y-axis 2, sin(x) graph
```

```python
# Using np.linspace()
x = np.linspace(0, 6, 100) #? HOW TO USE THIS INSTEAD OF SPECIFYING VALUES

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left') # For graph on y-axis 1, x2 graph

plt.twinx() # This creates a new y-axis for the plots that comes after

plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right') # For graph on y-axis 2, sin(x) graph
```

sin(x) graph shape is not smooth due to lack of plots

# 3 Saving to disc

```python
# saving image generated to computer using savefig()
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150) ###!!!
# Function accepts 'name.format' and dpi = resolution size
# Can replace 'name' with a specific location in your comp to save in
```
