#### Exercise 1

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             # marker = symbol/mark on each POINT of data
             # markerfacecolor = color of marker POINT
             ecolor='black',capsize=5,
             # ecolor = errorbar color
             label='Y values with errors')
             # adding "label=" after defining "marker" will produce a
             # name for the markers, which can be shown using plt.legend
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right') # shows legend for anything 
                              # that has been labelled in this graph?
plt.show()
```

```python
import numpy as np
import math
```

```python
x = np.array(list(range(0, 50)))
x = np.linspace(0, 50, 300)
                                # linspace can be used to generate
                                # a specified number of data points 
                                # within a given range, whereas the
                                # array will give as many data points 
                                # as the maximum limit of the range

                                # np.linspace(start, stop, num)
                                # where num = number of evenly spaced 
                                # values to be generated

y = np.exp(-x/10)*np.sin(x) # should i use np.linsoace for y too?

plt.plot(x, y, color  ='blue')

plt.ylim(-1, 1)
plt.grid(alpha = 0.25)
plt.xlabel('x', fontsize = 15, color = 'darkorange')
plt.ylabel('y', fontsize = 15, color = 'darkorange', rotation = 'horizontal')
plt.title(r'$Y = e^{-\frac{x}{10}}\sin(x)$', fontsize=20)
# plt.legend(loc='lower right') #! This line is not necessary as there
                                #! is no label for the plot (graph)
```
